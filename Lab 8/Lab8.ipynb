```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from tabulate import tabulate
from collections import Counter
```

```python
dataset = ["I love playing football on the weekends",
           "I enjoy hiking and camping in the mountains",
           "I like to read books and watch movies",
           "I prefer playing video games over sports",
           "I love listening to music and going to concerts"]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(dataset)
```

```python
k = 2 # Define the number of clusters
km = KMeans(n_clusters=k)
km.fit(X)

# Predict the clusters for each document
y_pred = km.predict(X)

# Display the document and its predicted cluster in a table
table_data = [["Document", "Predicted Cluster"]]
table_data.extend([[doc, cluster] for doc, cluster in zip(dataset, y_pred)])
print(tabulate(table_data, headers="firstrow"))

# Print top terms per cluster
print("\nTop terms per cluster:")
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()
for i in range(k):
    print("Cluster %d:" % i)
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind])
        
    print()
```

```python
# Calculate purity
total_samples = len(y_pred)
cluster_label_counts = [Counter(y_pred)]
purity = sum(max(cluster.values()) for cluster in cluster_label_counts) / total_samples

print("Purity:", purity)
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from tabulate import tabulate
from collections import Counter
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
dataset = [
    "I love playing football on the weekends",
    "I enjoy hiking and camping in the mountains",
    "I like to read books and watch movies",
    "I prefer playing video games over sports",
    "I love listening to music and going to concerts"
]
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

preprocessed_dataset = []
for doc in dataset:
    # Convert to lowercase
    doc = doc.lower()
    
    # Remove punctuation
    doc = doc.translate(str.maketrans('', '', string.punctuation))
    
    # Remove numbers
    doc = re.sub(r'\d+', '', doc)
    
    # Tokenize and remove stopwords, then stem
    words = doc.split()
    processed_words = []
    for word in words:
        if word not in stop_words:
            stemmed_word = stemmer.stem(word)
            processed_words.append(stemmed_word)
    
    # Join back into a string
    preprocessed_dataset.append(' '.join(processed_words))

print("Original documents vs Preprocessed documents:")
for orig, prep in zip(dataset, preprocessed_dataset):
    print(f"Original: {orig}")
    print(f"Preprocessed: {prep}")
    print()
```

```python
# Vectorize the preprocessed dataset
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(preprocessed_dataset)

# Perform clustering
k = 2
km = KMeans(n_clusters=k, random_state=42)
km.fit(X)

# Predict the clusters for each document
y_pred = km.predict(X)

# Display the document and its predicted cluster in a table
table_data = [["Document", "Predicted Cluster"]]
table_data.extend([[doc, cluster] for doc, cluster in zip(dataset, y_pred)])
print(tabulate(table_data, headers="firstrow"))

# Print top terms per cluster
print("\nTop terms per cluster:")
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()
for i in range(k):
    print(f"Cluster {i}:")
    for ind in order_centroids[i, :10]:
        print(f'  {terms[ind]}')
    print()
```

```python
# Calculate purity
total_samples = len(y_pred)

# Create a contingency table
clusters = y_pred

# For purity, we need actual labels (we don't have them, so this is approximate)
# Assuming we know the intended grouping: docs 0,1,3 are sports/outdoor? Actually let's manually assign
# Based on content: docs 0,1,3 seem active/sports, docs 2,4 are entertainment
true_labels = [0, 0, 1, 0, 1]  # 0=sports/active, 1=entertainment

# Calculate purity
cluster_label_counts = []
for cluster_id in range(k):
    mask = y_pred == cluster_id
    if np.any(mask):
        labels_in_cluster = np.array(true_labels)[mask]
        if len(labels_in_cluster) > 0:
            cluster_label_counts.append(Counter(labels_in_cluster))

purity = sum(max(counts.values()) for counts in cluster_label_counts) / total_samples
print(f"Purity: {purity}")
```

```python
dataset = [
    "I love playing football on the weekends",
    "I enjoy hiking and camping in the mountains",
    "I like to read books and watch movies",
    "I prefer playing video games over sports",
    "I love listening to music and going to concerts"
]
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

preprocessed_dataset = []
for doc in dataset:
    # Convert to lowercase
    doc = doc.lower()
    
    # Remove punctuation
    doc = doc.translate(str.maketrans('', '', string.punctuation))
    
    # Remove numbers
    doc = re.sub(r'\d+', '', doc)
    
    # Tokenize and remove stopwords, then stem
    words = doc.split()
    processed_words = []
    for word in words:
        if word not in stop_words:
            stemmed_word = stemmer.stem(word)
            processed_words.append(stemmed_word)
    
    preprocessed_dataset.append(processed_words)

print("Preprocessed tokens for Word2Vec:")
for i, tokens in enumerate(preprocessed_dataset):
    print(f"Document {i}: {tokens}")
print()
```

```python
# Train Word2Vec model on preprocessed data
word2vec_model = Word2Vec(
    sentences=preprocessed_dataset, 
    vector_size=100, 
    window=5, 
    min_count=1, 
    workers=4,
    seed=42
)
```

```python
# Create document embeddings
X = []
for doc_tokens in preprocessed_dataset:
    word_vectors = []
    for word in doc_tokens:
        if word in word2vec_model.wv:
            word_vectors.append(word2vec_model.wv[word])
    if word_vectors:
        doc_vector = np.mean(word_vectors, axis=0)
    else:
        doc_vector = np.zeros(100)
    X.append(doc_vector)

X = np.array(X)
```

```python
# Perform clustering
k = 2
km = KMeans(n_clusters=k, random_state=42)
km.fit(X)

# Predict the clusters for each document
y_pred = km.predict(X)

# Tabulate the document and predicted cluster
table_data = [["Document", "Predicted Cluster"]]
table_data.extend([[doc, cluster] for doc, cluster in zip(dataset, y_pred)])
print(tabulate(table_data, headers="firstrow"))
```

```python
# Calculate purity
total_samples = len(y_pred)
true_labels = [0, 0, 1, 0, 1]  # 0=sports/active, 1=entertainment

cluster_label_counts = []
for cluster_id in range(k):
    mask = y_pred == cluster_id
    if np.any(mask):
        labels_in_cluster = np.array(true_labels)[mask]
        if len(labels_in_cluster) > 0:
            cluster_label_counts.append(Counter(labels_in_cluster))

purity = sum(max(counts.values()) for counts in cluster_label_counts) / total_samples
print(f"\nPurity: {purity}")
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from tabulate import tabulate
from collections import Counter
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
df = pd.read_csv('customer_complaints_1.csv')
```

```python
print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst few rows:")
print(df.head())
print("\n" + "="*80 + "\n")
```

```python
complaints = df['text'].fillna('').tolist()

print(f"Number of complaints: {len(complaints)}")
print(f"Sample complaint:\n{complaints[0][:500]}...")
print("\n" + "="*80 + "\n")
```

```python
# Text preprocessing for complaints
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

preprocessed_complaints = []
original_lengths = []
processed_lengths = []

for complaint in complaints:
    # Store original length for comparison
    original_lengths.append(len(complaint))
    
    # Convert to lowercase
    complaint = complaint.lower()
    
    # Remove punctuation
    complaint = complaint.translate(str.maketrans('', '', string.punctuation))
    
    # Remove numbers
    complaint = re.sub(r'\d+', '', complaint)
    
    # Remove extra whitespace
    complaint = ' '.join(complaint.split())
    
    # Tokenize and remove stopwords, then stem
    words = complaint.split()
    processed_words = []
    for word in words:
        if word not in stop_words and len(word) > 2:
            stemmed_word = stemmer.stem(word)
            processed_words.append(stemmed_word)
    
    # Join back into a string
    preprocessed_complaint = ' '.join(processed_words)
    preprocessed_complaints.append(preprocessed_complaint)
    processed_lengths.append(len(preprocessed_complaint))

# Show preprocessing impact on first complaint
print("Preprocessing Example:")
print(f"Original (first 500 chars):\n{complaints[0][:500]}")
print(f"\nPreprocessed:\n{preprocessed_complaints[0][:500]}")
print(f"\nOriginal length: {original_lengths[0]} chars")
print(f"Preprocessed length: {processed_lengths[0]} chars")
print(f"Reduction: {(1 - processed_lengths[0]/original_lengths[0])*100:.1f}%")
print("\n" + "="*80 + "\n")
```

```python
# Vectorize using TF-IDF
vectorizer = TfidfVectorizer(max_features=1000, min_df=2, max_df=0.8)
X = vectorizer.fit_transform(preprocessed_complaints)
print(f"TF-IDF matrix shape: {X.shape}")
print(f"Number of features (terms): {len(vectorizer.get_feature_names_out())}")
print("\n" + "="*80 + "\n")
```

```python
# Find optimal number of clusters using elbow method
print("Finding optimal number of clusters...")
inertias = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)
    print(f"  K={k}: Inertia={km.inertia_:.2f}")

# Calculate elbow point (simple method)
if len(inertias) > 1:
    diffs = np.diff(inertias)
    elbow_k = K_range[np.argmin(diffs) + 1]
    print(f"\nSuggested optimal K: {elbow_k}")
else:
    elbow_k = 3
```

```python
# Perform clustering with chosen K
k = elbow_k  # Use optimal K from elbow method
print(f"Performing K-means clustering with K={k}...")
km = KMeans(n_clusters=k, random_state=42, n_init=10)
km.fit(X)

# Predict clusters
y_pred = km.predict(X)
df['cluster'] = y_pred
```

```python
# Display results
print("\nCluster distribution:")
cluster_counts = Counter(y_pred)
for cluster_id in range(k):
    print(f"  Cluster {cluster_id}: {cluster_counts[cluster_id]} complaints ({cluster_counts[cluster_id]/len(y_pred)*100:.1f}%)")

# Display sample complaints from each cluster
print("\n" + "="*80)
print("SAMPLE COMPLAINTS FROM EACH CLUSTER:")
print("="*80)

for cluster_id in range(k):
    print(f"\n--- CLUSTER {cluster_id} ---")
    cluster_indices = np.where(y_pred == cluster_id)[0]
    print(f"Total complaints in cluster: {len(cluster_indices)}")
    
    # Show first 3 complaints from this cluster
    print("Sample complaints:")
    for idx in cluster_indices[:3]:
        original_text = complaints[idx]
        # Truncate long texts
        if len(original_text) > 200:
            original_text = original_text[:200] + "..."
        print(f"  • {original_text}")
    
    # Show top terms for this cluster
    print(f"\nTop terms for cluster {cluster_id}:")
    cluster_center = km.cluster_centers_[cluster_id]
    top_term_indices = cluster_center.argsort()[-10:][::-1]
    terms = vectorizer.get_feature_names_out()
    top_terms = [terms[idx] for idx in top_term_indices]
    print(f"  {', '.join(top_terms)}")
```

```python
# Display full results table (first 20 complaints)
print("CLUSTERING RESULTS (First 20 complaints):")
table_data = [["Index", "Author", "Rating", "Cluster", "Complaint (truncated)"]]
for i in range(min(20, len(df))):
    complaint_short = df.iloc[i]['text'][:100] + "..." if len(df.iloc[i]['text']) > 100 else df.iloc[i]['text']
    table_data.append([
        i, 
        df.iloc[i]['author'][:30], 
        df.iloc[i]['rating'], 
        df.iloc[i]['cluster'],
        complaint_short
    ])
print(tabulate(table_data, headers="firstrow", tablefmt="grid"))
```

```python
# Step 9: Analyze clusters by rating
print("CLUSTER ANALYSIS BY RATING:")
for cluster_id in range(k):
    cluster_data = df[df['cluster'] == cluster_id]
    rating_dist = cluster_data['rating'].value_counts().sort_index()
    print(f"\nCluster {cluster_id}:")
    print(f"  Average rating: {cluster_data['rating'].mean():.2f}")
    print(f"  Rating distribution:")
    for rating, count in rating_dist.items():
        print(f"    {rating} star: {count} complaints ({count/len(cluster_data)*100:.1f}%)")
    print(f"  Primary issues: ", end="")
    # Get top terms for this cluster
    cluster_center = km.cluster_centers_[cluster_id]
    top_term_indices = cluster_center.argsort()[-5:][::-1]
    terms = vectorizer.get_feature_names_out()
    top_terms = [terms[idx] for idx in top_term_indices]
    print(f"{', '.join(top_terms)}")
```
